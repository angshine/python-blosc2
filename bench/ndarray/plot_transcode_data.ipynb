```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_parquet("measurements-01369-clevels.parquet")
df = df.query("clevel > 0")  # get rid of no compression results
```

```python
hover_data = {"filter": False, "codec": True, "cratio": ':.1f', "cspeed": ':.1f', "dspeed": ':.1f', "dset": False, "clevel": True}
fig = px.violin(df, y="cratio", x="dset", color="filter", points="all", hover_data=hover_data)
fig.show()
```

```python
hover_data = {"filter": False, "codec": False, "cratio": ':.1f', "cspeed": ':.1f', "dspeed": ':.1f', "dset": True, "clevel": True}
fig = px.violin(df, y="cratio", x="codec", color="filter", points="all", hover_data=hover_data)
fig.show()
```
